```python
import pandas as pd
train_df_prev = pd.read_csv("train.csv")
dev_df_prev = pd.read_csv("dev.csv")
test_df_prev = pd.read_csv("test.csv")
```

```python
train_df_prev.head(5)
```

```python
train_df_prev.columns
```

```python
train_df = train_df_prev.filter(['topic', 'argument', 'key_point', 'label'], axis=1)
dev_df = dev_df_prev.filter(['topic', 'argument', 'key_point', 'label'], axis=1)
test_df = test_df_prev.filter(['topic', 'argument', 'key_point', 'label'], axis=1)
```

```python
train_df.head(5)
```

```python
# No Prompt Engineering

# For Train set

for i in train_df.index:
  topic = train_df['topic'][i]
  arg = train_df['argument'][i]
  key = train_df['key_point'][i]
  if arg[-1] != '.':
    modified_arg = topic + '. ' + arg + '. ' + key + '.'
    train_df.at[i, 'argument'] = modified_arg
  else:
    modified_arg = topic + '. ' + arg + ' ' + key + '.'
    train_df.at[i, 'argument'] = modified_arg

# For Dev set

for i in dev_df.index:
  topic = dev_df['topic'][i]
  arg = dev_df['argument'][i]
  key = dev_df['key_point'][i]
  if arg[-1] != '.':
    modified_arg = topic + '. ' + arg + '. ' + key + '.'
    dev_df.at[i, 'argument'] = modified_arg
  else:
    modified_arg = topic + '. ' + arg + ' ' + key + '.'
    dev_df.at[i, 'argument'] = modified_arg

# For Test set

for i in test_df.index:
  topic = test_df['topic'][i]
  arg = test_df['argument'][i]
  key = test_df['key_point'][i]
  if arg[-1] != '.':
    modified_arg = topic + '. ' + arg + '. ' + key + '.'
    test_df.at[i, 'argument'] = modified_arg
  else:
    modified_arg = topic + '. ' + arg + ' ' + key + '.'
    test_df.at[i, 'argument'] = modified_arg
```

```python
test_df.at[110, 'argument']
```

```python
train_df = train_df.filter(['argument', 'label'], axis=1)
dev_df = dev_df.filter(['argument', 'label'], axis=1)
test_df = test_df.filter(['argument', 'label'], axis=1)
```

```python
train_df["label"] = train_df["label"].apply(str)
dev_df["label"] = dev_df["label"].apply(str)
test_df["label"] = test_df["label"].apply(str)
```

```python
# Renaming the columns
train_df.columns = ["input_text","target_text"]
dev_df.columns = ["input_text","target_text"]
test_df.columns = ["input_text","target_text"]
```

```python
# Adding a prefix. Here we shall keep "paraphrase" as a prefix.
train_df["prefix"] = "text-classification"
dev_df["prefix"] = "text-classification"
test_df["prefix"] = "text-classification"
```

```python
from sklearn.model_selection import train_test_split
train_data, dev_data = train_df, dev_df
```

```python
args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "max_seq_length": 256,
    "num_train_epochs": 4,
    "num_beams": None,
    "do_sample": True,
    "top_k": 50,
    "top_p": 0.95,
    "use_multiprocessing": False,
    "save_steps": -1,
    "save_eval_checkpoints": False,
    "evaluate_during_training": False,
    "adam_epsilon": 1e-08,
    "eval_batch_size": 6,
    "fp_16": False,
    "gradient_accumulation_steps": 16,
    "learning_rate": 0.0003,
    "max_grad_norm": 1.0,
    "n_gpu": 1,
    "seed": 42,
    "train_batch_size": 6,
    "warmup_steps": 0,
    "weight_decay": 0.0,
}
```

```python
from simpletransformers.t5 import T5Model, T5Args
import sklearn

#model_args = T5Args()
#model_args.num_train_epochs = 4

model = T5Model("t5","t5-small", args=args)
```

```python
model.train_model(train_data, eval_data=dev_data, use_cuda=True,acc=sklearn.metrics.accuracy_score)
```

```python
from simpletransformers.t5 import T5Model
import os
root_dir = os.getcwd()
trained_model_path = os.path.join(root_dir,"/content/outputs")
args = {
"overwrite_output_dir": True,
"max_seq_length": 256,
"max_length": 50,
"top_k": 50,
"top_p": 0.95,
"num_return_sequences": 1
}
trained_model = T5Model("t5",trained_model_path,args=args)
```

```python
# Predict an input

prefix = "text-classification"
pred = trained_model.predict([f"{prefix}: Under austerity regimes, public services and other societal benefits are reduced. Austerity cuts access to essential services."])
print(pred)
```

```python
# Predict full test set

prefix = "text-classification"
ref = []
to_predict = []
for i in test_df.index:
  input = prefix + ": " + test_df['input_text'][i]
  ref.append(test_df['target_text'][i])
  to_predict.append(input)
  #print(input)
  #pred.append(trained_model.predict(f"{prefix}: {input}"))

predictions = trained_model.predict(to_predict)
```

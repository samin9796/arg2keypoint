```python
from transformers import BartTokenizer, BartForConditionalGeneration

model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
```

```python
ARTICLE_TO_SUMMARIZE = (
    "It's official: U.S. President Barack Obama wants lawmakers to weigh in on whether to use military force in Syria."
    "Obama sent a letter to the heads of the House and Senate on Saturday night, "
    "hours after announcing that he believes military action against Syrian targets is the right step to take "
    "over the alleged use of chemical weapons." 
    "The proposed legislation from Obama asks Congress to approve the use of military force "
    "to deter, disrupt, prevent and degrade the potential for future uses of chemical weapons or other weapons of mass destruction."
)
```

```python
inputs = tokenizer([ARTICLE_TO_SUMMARIZE], max_length=1024, return_tensors="pt")
```

```python
# Generate Summary
summary_ids = model.generate(inputs["input_ids"], num_beams=2, max_length=20)
tokenizer.batch_decode(summary_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]
```

```python
import torch
device = torch.device('cpu')

text = "I know what you told about the matter when the bell was ringing and you were at the place which is called England."


preprocess_text = text.strip().replace("\n","")
t5_prepared_Text = "summarize: "+preprocess_text
print ("original text preprocessed: \n", preprocess_text)

tokenized_text = tokenizer.encode(t5_prepared_Text, return_tensors="pt").to(device)

summary_ids = model.generate(tokenized_text,
                        num_beams=10,
                        no_repeat_ngram_size=1,
                        min_length=10,
                        num_return_sequences = 2,
                        max_length=20,
                        top_k = 100,
                        early_stopping=True)

output = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
output1 = tokenizer.decode(summary_ids[1], skip_special_tokens=True)
```

```python
import pandas as pd
dataset_df = pd.read_csv("ArgKP_dataset.csv")
```

```python
dataset_df.head
```

```python
dataset_df.columns
```

```python
dataset_df_new = dataset_df.filter(['argument', 'key_point'], axis=1)
```

```python
dataset_df_new.head
```

```python
# Renaming the columns
dataset_df_new.columns = ["input_text","target_text"]
```

```python
# Adding a prefix. Here we shall keep "paraphrase" as a prefix.
dataset_df_new["prefix"] = "paraphrase"
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data = train_test_split(dataset_df_new,test_size=0.1)
```

```python
args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "max_seq_length": 256,
    "num_train_epochs": 4,
    "num_beams": None,
    "do_sample": True,
    "top_k": 50,
    "top_p": 0.95,
    "use_multiprocessing": False,
    "save_steps": -1,
    "save_eval_checkpoints": True,
    "evaluate_during_training": False,
    "adam_epsilon": 1e-08,
    "eval_batch_size": 6,
    "fp_16": False,
    "gradient_accumulation_steps": 16,
    "learning_rate": 0.0003,
    "max_grad_norm": 1.0,
    "n_gpu": 1,
    "seed": 42,
    "train_batch_size": 6,
    "warmup_steps": 0,
    "weight_decay": 0.0,
}
```

```python
from simpletransformers.t5 import T5Model
import sklearn
model = T5Model("t5","t5-small", args=args)
```

```python
model.train_model(train_data, eval_data=test_data, use_cuda=True,acc=sklearn.metrics.accuracy_score)
```
